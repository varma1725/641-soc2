```python
from bs4 import BeautifulSoup
import requests

url="https://crawler-test.com/"

response=requests.get(url)
print('The status code is',response.status_code)
```

```python
#to access first 100 char
print(response.text[:100])
```

```python
#to get the title
#to access the site
soup=BeautifulSoup(response.text,'html.parser')
print(soup.find('title').text)
```

```python
heading=soup.find('h1')
print(heading.text)
```

```python
#to find the tag
links=soup.find('a')
print(links)
```

```python
#to print all links
all_links=soup.find_all('a')
print(type(all_links))
print('----------------------------')
for val in all_links:
    print(val)
    print()
```

```python
#to find element by ID
head=soup.find(id='header')
print(head)
```

```python
#to access a tag in particular ID
a=head.find('a')
print(a)
```

```python
# find the element based on class
class_based=soup.find(class_='row side-collapsed')
print(class_based)
```

```python
class_based=soup.find(class_='panel-header')
print(class_based.text)
```

```python
headings =soup.find_all('div',{'class':'panel-header'})
for val in headings:
    h3 = val.find('h3')
    print(h3.text)
```

```python
s=soup.find_all('h3')
for i in s:
    print(i.text)
```
