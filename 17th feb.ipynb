```python
# k means clustering

# K-means clustering is a popular unsupervised machine learning algorithm used for partitioning a 
# dataset into K distinct, non-overlapping subsets (clusters). The goal is to group similar data points 
# together and identify underlying patterns in the data. Here's a brief overview of how K-means 
# clustering works:
```

```python
from sklearn.datasets import make_blobs
#The make_blob functions isn used to generate synthetic data sets for clustering and classification
#tasks.
#This function will create clustures of data points with gaussian distribution
```

```python
#create a random data set
ds=make_blobs(n_samples=200,n_features=2,centers=5,cluster_std=1.5,random_state=101)
#n_samples=Total no.of points equally divided among clusters.
#n_features=It indicate the no of features(columns)
#centers=it determine the no of clusters to be generated
#luster_std= it set the standard deviation. theclusters to spread out
```

```python
ds
```

```python
import matplotlib.pyplot as plt
x,y=ds
plt.scatter(x[:,0],x[:,1],c=y,cmap='rainbow',edgecolor='black',s=100)
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('Scatter plot for k means')
```

```python
ds[0].shape
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=4)
```

```python
km.fit(ds[0])
```

```python
km.cluster_centers_
```

```python
fig, (ax1,ax2)=plt.subplots(1,2,figsize=(10,6))
ax1.set_title('k means') #predicted data
ax1.scatter(ds[0][:,0],ds[0][:,1],c=km.labels_,cmap='rainbow',edgecolor="black")
ax2.set_title('Original') #original data
ax2.scatter(ds[0][:,0],ds[0][:,1],c=ds[1],cmap='rainbow',edgecolor="black")
```
